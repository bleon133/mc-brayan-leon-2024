```python
#Brayan Steven León Martinez - Taller 19 - Matemáticas Computacional
import math
from math import e
import numpy as np
import matplotlib.pyplot as plt
#Variables
x=[1,2,3,4,5,6,7,8]
y=[4.3,6.5,7.5,8,8.5,8.8,9,9.5]
n=len(x)
```

```python
plt.plot(x, y, 'ro')
plt.show()
```

```python
#regresion lineal 1
multxyReg1 = []
highxReg1 = []
#regresion lineal 2
lnyReg2=[]
multxlnyReg2=[]
highxReg2 = []
#potencias
lnxPot=[]
lnyPot=[]
multlnxlnyPot=[]
highlnxPot = []
#razón
divxRaz=[]
divyRaz=[]
multdivxdivyRaz=[]
highdivxRaz=[]
for i in range(n):
  #regresion lineal 1
  multxyReg1.append(x[i]*y[i])
  highxReg1.append(x[i]**2)
  #regresion lineal 2
  lnyReg2.append(np.log(y[i]))
  multxlnyReg2.append(x[i]*lnyReg2[i])
  highxReg2.append(x[i]**2)
  #potencias
  lnxPot.append(math.log10(x[i]))
  lnyPot.append(math.log10(y[i]))
  multlnxlnyPot.append(lnxPot[i]*lnyPot[i])
  highlnxPot.append(lnxPot[i]**2)
  #razón
  divxRaz.append(1/x[i])
  divyRaz.append(1/y[i])
  multdivxdivyRaz.append(divxRaz[i]*divyRaz[i])
  highdivxRaz.append(divxRaz[i]**2)
```

```python
#regresion lineal 1
sumatoria_xReg1=sum(x)
sumatoria_yReg1=sum(y)
sumatoria_multxyReg1=sum(multxyReg1)
sumatoria_highxReg1=sum(highxReg1)
#regresion lineal 2
sumatoria_xReg2=sum(x)
sumatoria_lnyReg2=sum(lnyReg2)
sumatoria_multxlnyReg2=sum(multxlnyReg2)
sumatoria_highxReg2=sum(highxReg2)
#potencias
sumatoria_lnxPot=sum(lnxPot)
sumatoria_lnyPot=sum(lnyPot)
sumatoria_multlnxlnyPot=sum(multlnxlnyPot)
sumatoria_highlnxPot=sum(highlnxPot)
#razón
sumatoria_divxRaz=sum(divxRaz)
sumatoria_divyRaz=sum(divyRaz)
sumatoria_multdivxdivyRaz=sum(multdivxdivyRaz)
sumatoria_highdivxRaz=sum(highdivxRaz)
```

```python
#regresion lineal 1
promedio_xReg1=sumatoria_xReg1/n
promedio_yReg1=sumatoria_yReg1/n
#regresion lineal 2
promedio_xReg2=sumatoria_xReg2/n
promedio_lnyReg2=sumatoria_lnyReg2/n
#potencias
promedio_lnxPot=sumatoria_lnxPot/n
promedio_lnyPot=sumatoria_lnyPot/n
#razón
promedio_DivxRaz=sumatoria_divxRaz/n
promedio_DivyRaz=sumatoria_divyRaz/n
```

```python
#regresion lineal 1
a1Reg1=((n*sumatoria_multxyReg1)-(sumatoria_xReg1*sumatoria_yReg1))/((n*sumatoria_highxReg1-sumatoria_xReg1**2))
a0Reg1=promedio_yReg1-(a1Reg1*promedio_xReg1)
#regresion lineal 2
a1Reg2=((n*sumatoria_multxlnyReg2)-(sumatoria_xReg2*sumatoria_lnyReg2))/((n*sumatoria_highxReg2-sumatoria_xReg2**2))
a0Reg2=promedio_lnyReg2-(a1Reg2*promedio_xReg2)
#potencias
a1Pot=((n*sumatoria_multlnxlnyPot)-(sumatoria_lnxPot*sumatoria_lnyPot))/((n*sumatoria_highlnxPot-sumatoria_lnxPot**2))
a0Pot=promedio_lnyPot-(a1Pot*promedio_lnxPot)
#razón
a1Raz=((n*sumatoria_multdivxdivyRaz)-(sumatoria_divxRaz*sumatoria_divyRaz))/((n*sumatoria_highdivxRaz-sumatoria_divxRaz**2))
a0Raz=promedio_DivyRaz-(a1Raz*promedio_DivxRaz)
```

```python
#regresion lineal 1
print("y = " + str(a0Reg1) + " + " + str(a1Reg1) + "x")
#regresion lineal 2
alfaReg2 = e**a0Reg2
betaReg2 = a1Reg2
print("y = " + str(alfaReg2) + "*e^" + str(betaReg2) + "x")
#potencias
alfaPot = 10**a0Pot
betaPot = a1Pot
print("y = " + str(alfaPot) + "*x^" + str(betaPot))
#razón
alfaRaz = 1/a0Raz
betaRaz = a1Raz*alfaRaz
print("y = " + str(alfaRaz) + "*" + "(x/" + str(betaRaz) + "+x)")
```

```python
#regresion lineal 1
ryReg1=[]
for i in range(n):
    ryReg1.append(a0Reg1 + (a1Reg1*x[i]))
#regresion lineal 2
ryReg2=[]
for i in range(n):
    ryReg2.append(alfaReg2*(e**(betaReg2*x[i])))
#potencias
ryPot=[]
for i in range(n):
    ryPot.append(alfaPot*(x[i]**betaPot))
#razón
ryRaz=[]
for i in range(n):
    ryRaz.append(alfaRaz*(x[i]/(betaRaz+x[i])))
```

```python
#regresion lineal 1
plt.plot(x, y, 'r--', x, ryReg1, 'bs')
plt.show()
```

```python
#regresion lineal 2
plt.plot(x, y, 'r--', x, ryReg2, 'bs')
plt.show()
```
